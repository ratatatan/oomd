Вова подбрасывает два шестигранных кубика, а Лёша записывает получившуюся сумму. Вова подбросывал кубики примерно 20 лет, и за это время подбросил их ровно 50 миллиардов раз. И ещё столько же подбросит.

Для задания для своего варианта запустите следующую ячейку:

```python
if (int(str(v)[5]) % 2) == 0:
  print ('Напишите программу, которая выводит три самые частые суммы, упорядоченные по убыванию частот.')
else:
  print ('Напишите программу, которая выводит пять самых редких сумм, упорядоченных по возрастанию частот.')
```

Вывод должен включать в себя сумму и указание количество выпадений этой суммы.

Ниже описан пример генератора случайных чисел

В этом примере кубик всего один.

```python
import random
```

```python
random.seed(4242)
x = []
for i in range(10):
  x.append(random.randint(1, 6))
print (x)
```

В этом примере. Грань 1 выпала 1 раз, 2 выпала 2 раза, 3 выпало 2 раза, 4 выпало 3 раза, 5 выпало 2 раза, 6 выпало 0 раз

```python
random.seed(4242)
x = {(i, j): 0 for i in range(1, 7) for j in range(1, 7)}
for i in range(50*1000_000_000):
  x[(random.randint(1, 6), random.randint(1, 6))] += 1

answer = sorted(x, key=lambda k: x[k])[-3:]

for _ in range(3):
  print(f"{answer[_]}: {x[answer[_]]}")
```
